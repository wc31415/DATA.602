```python
#run the following command in a command prompt first:
#ipcluster start --n=4

from Tkinter import Tk
import tkFileDialog
from random import choice
from IPython import parallel as p
import timeit

root = Tk()
root.withdraw()


class AppleStockPrice:
    count = 0
    def __init__(self, input_date, input_price, input_percent_change):
        self.__class__.count += 1
        self.date = input_date
        self.price = input_price
        self.percent_change = input_percent_change
        

#Takes 2 parameters:
#-list of percent changes in Apple stock
#-last price of Apple stock
#randomly generates 20 price changes
#takes 20 price changes and calculates a predicted stock price
#returns predicted stock price
def predictedPrice(apple_percent_change, apple_last_price):
    apple_percent_change_rng = []
    
    for i in range(20):
        apple_percent_change_rng.append(choice(apple_percent_change))
        apple_last_price = (apple_last_price * apple_percent_change_rng[i]) + apple_last_price        
    
    return apple_last_price


#Takes 3 parameters:
#-list of percent changes in Apple stock
#-last price of Apple stock
#-empty list to contain predicted stock prices
#returns list of predicted stock prices
def parallelMC(apple_percent_change, apple_last_price, apple_predicted_prices):
    for i in range(3750):
        temp = predictedPrice(apple_percent_change, apple_last_price)
        apple_predicted_prices.append(temp)
        
    return apple_predicted_prices


def timeit_serialMC(apple_percent_change, apple_last_price):
    apple_predicted_prices = []
        
    for i in range(15000):
        temp = predictedPrice(apple_percent_change, apple_last_price)
        apple_predicted_prices.append(temp)
        
        
def timeit_parallelMC(apple_percent_change, apple_last_price, apple_predicted_prices):
    for i in range(3750):
        temp = predictedPrice(apple_percent_change, apple_last_price)
        apple_predicted_prices.append(temp)
            
        
if __name__ == "__main__":
    filename = tkFileDialog.askopenfilename()    
    f = open(filename)
    apple_2011 = []
    for line in f:
        line = line.strip()
        input_list = line.split(",")
        apple_2011.append(AppleStockPrice(input_list[0], input_list[1], input_list[2]))
        
    apple_percent_change = []
    
    for i in range(2, AppleStockPrice.count):        
        apple_percent_change.append(float(apple_2011[i].percent_change))
        
    apple_last_price = float(apple_2011[AppleStockPrice.count - 1].price)
    
    apple_predicted_prices = []
           
    rc = p.Client()
    
    rc.block = True
    
    print "Check - 4 engines"
    print rc.ids
        
    rc[:].execute("from random import choice")
    rc[:]['predictedPrice'] = predictedPrice
    
    parallel_results = rc[:].apply(parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices)
    
    parallel_results_combined = []
    
    map(parallel_results_combined.extend, parallel_results)
    
    parallel_results_combined.sort()
    
    print ("\nMonte Carlo simulation to calculate Value at Risk (VaR)\n" \
           "for Apple stock - 15,000 total simulations\n" \
           "Completed in parallel using 4 engines\n" \
           "3750 simulations per engine\n" \
           "Bottom 1% of predicted Apple stock prices displayed\n" \
           "Total: 150 predicted prices\n")
    
    for i in range(150):
        print parallel_results_combined[i]
       

    print "\ntimeit"
    
    print ("\nHW 10 Version - Monte Carlo simulation\n" \
           "15,000 simulations performed in serial - 10 loops\n")
    
    print str((timeit.timeit(stmt = "timeit_serialMC(apple_percent_change, apple_last_price)",\
                             setup = "from __main__ import timeit_serialMC, apple_percent_change, apple_last_price",\
                             number = 10))) + " seconds"
    
    print ("\nParallel Computing Version - Monte Carlo simulation\n" \
           "15,000 simulations performed in parallel\n" \
           "Split between 4 engines - 10 loops\n")
    
    print str((timeit.timeit(stmt = "rc[:].apply(timeit_parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices)",\
                             setup = "from __main__ import rc, timeit_parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices",\
                             number = 10))) + " seconds"
    
    print ("\nHW 10 Version - Monte Carlo simulation\n" \
           "15,000 simulations performed in serial - 100 loops\n")
    
    print str((timeit.timeit(stmt = "timeit_serialMC(apple_percent_change, apple_last_price)",\
                             setup = "from __main__ import timeit_serialMC, apple_percent_change, apple_last_price",\
                             number = 100))) + " seconds"
    
    print ("\nParallel Computing Version - Monte Carlo simulation\n" \
           "15,000 simulations performed in parallel\n" \
           "Split between 4 engines - 100 loops\n")
    
    print str((timeit.timeit(stmt = "rc[:].apply(timeit_parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices)",\
                             setup = "from __main__ import rc, timeit_parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices",\
                             number = 100))) + " seconds"
    
    print ("\nHW 10 Version - Monte Carlo simulation\n" \
           "15,000 simulations performed in serial - 500 loops\n")
    
    print str((timeit.timeit(stmt = "timeit_serialMC(apple_percent_change, apple_last_price)",\
                             setup = "from __main__ import timeit_serialMC, apple_percent_change, apple_last_price",\
                             number = 500))) + " seconds"
    
    print ("\nParallel Computing Version - Monte Carlo simulation\n" \
           "15,000 simulations performed in parallel\n" \
           "Split between 4 engines - 500 loops\n")
    
    print str((timeit.timeit(stmt = "rc[:].apply(timeit_parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices)",\
                             setup = "from __main__ import rc, timeit_parallelMC, apple_percent_change, apple_last_price, apple_predicted_prices",\
                             number = 500))) + " seconds"
```
